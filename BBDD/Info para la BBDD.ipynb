```python
import pandas as pd
import numpy as np
import re
```

```python
bd  = pd.read_csv("datos_limpios_final.csv", index_col = 0)
pd.set_option("display.max.columns", None)
bd
```

```python
employees = ['age',
            'gender', 
            'marital_status']

employee_details = ['education',
                    'total_working_years',
                    'num_companies_worked']

company_relationship =['attrition',
                       'years_at_company',
                       'years_since_last_promotion',
                       'years_with_curr_manager']

satisfaction = ['environment_satisfaction',
                'relationship_satisfaction',
                'job_satisfaction',
                'job_involvement',
                'work_life_balance',
                'performance_rating']

job_role = ['job_role',
            'job_level',
            'training_times_last_year',
            'stock_option_level']

conditions = ['remote_work',
              'distance_from_home']

salary = ['hourly_rate',
          'monthly_rate',
          'percent_salary_hike']
```

```python
bd.isnull().sum()
```

```python
bd_employees = bd[employees]
bd_employee_details = bd[employee_details]
bd_company_relationship = bd[company_relationship]
bd_satisfaction = bd[satisfaction]
bd_job_role = bd[job_role]
bd_conditions = bd[conditions]
bd_salary = bd[salary]
```

```python
bd_employees.to_csv('employees.csv')
bd_employee_details.to_csv('employee_details.csv')
bd_company_relationship.to_csv('company_relationship.csv')
bd_satisfaction.to_csv('satisfaction.csv')
bd_job_role.to_csv('job_role.csv')
bd_conditions.to_csv('conditions.csv')
bd_salary.to_csv('salary.csv')
```
